```python
from sklearn.impute import SimpleImputer
```

```python
fords=pd.read_csv('fords.csv')
fords = fords[['Year','Mileage','Price','Age']]
```

```python
imr_mean = SimpleImputer(strategy='mean')
imr_mean = imr_mean.fit(fords[['Mileage']])
fords['Mileage'] = imr_mean.transform(fords[['Mileage']]).ravel()
fords.info()
```
